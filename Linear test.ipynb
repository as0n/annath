```python
from numeric import Numeric
import matplotlib.pyplot as plt
import numpy as np
```

```python
nn = Numeric(8, 16)
f = lambda x: x
```

```python
values = [0, 1, 2, 4, 8, 9, 16, 32, 64, 128]
```

```python
for v in values:
    nn.learn(v, f(v))
```

```python
nn.train(300000)
```

```python
nn.error()
```

```python
x = np.linspace(1, 255, 500)
plt.plot(x, f(x))
plt.plot(x, nn.apply_multiple(x))
plt.title('Estimation');
```
